```python
from keras import Sequential, models,optimizers
from keras.layers import *
import numpy as np
```

```python
Xs = np.array([x for x in range(1, 1001)])
Ys = []
for y in range(1, 1001):
  if y%2==1:
    Ys.append(0)
  else:
    Ys.append(1)
Ys = np.array(Ys)
```

```python
model = Sequential([
   Dense(1, activation=Activation('sigmoid'))                                   
])
```

```python
adam = optimizers.Adam(lr=1)
model.compile(optimizer=adam, loss="binary_crossentropy")
```

```python
model.fit(Xs, Ys, batch_size=10, epochs=1500)
```

```python
model.predict([99])
```
