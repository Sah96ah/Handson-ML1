```python
from tensorflow import keras
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sample_Projects/Datasets/IowaHousingPrices.csv')
ds.head()
```

```python
Xs = ds['SquareFeet']
Ys = ds['SalePrice']
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, activation='linear', input_shape=(1,))
])
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=1), loss='mean_squared_error')
```

```python
model.fit(Xs, Ys, batch_size=10, epochs=30, validation_split=0.2)
```

```python
print(model.predict([2000]))
```
