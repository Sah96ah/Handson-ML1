```python
from tensorflow.keras import Sequential, optimizers
from keras import backend as K
from keras.utils.generic_utils import get_custom_objects
from keras.models import load_model
from keras.layers import *
import numpy as np
```

```python
Xs = np.array([x for x in range(1, 101)])
Ys = np.array(Xs**2)
```

```python
#my custom activation function
def square(x):
  return K.square(x)

get_custom_objects().update({'square': Activation(square)})
```

```python
model = Sequential()
model.add(Dense(1, activation=square))
```

```python
model.compile(optimizers.Adam(learning_rate=1), 'mean_squared_error')
```

```python
model.fit(Xs,Ys, batch_size=10, epochs=1000, verbose=0)
```

```python
value = 101
pred = np.round(model.predict([value])[0][0])
pred
```
