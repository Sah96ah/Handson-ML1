```python
import numpy as np
from numpy import loadtxt
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import model_from_json
from sklearn.model_selection import train_test_split
```

```python
dataset = loadtxt('Data/pima-indians-diabetes.csv', delimiter=',')
X = dataset[:,0:8]
y = dataset[:,8]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=8)
```

# Evaluation

```python
_, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Accuracy is {accuracy * 100:.2f}')
```

```python
model_json = model.to_json()
with open ("Data/008/model.json", "w") as json_file:
  json_file.write(model_json)
model.save("Data/008/model.h5")
```

# Prediction

```python
json_file = open('Data/008/model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()


model = model_from_json(loaded_model_json)
model.load_weights("Data/008/model.h5")
```

```python
prediction = model.predict(X_test[10].reshape(-1, 8))
print(f'original class {int(y_test[5])}')
print(f'classified class {np.argmax(prediction)}')
```
