```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
Xs = np.array([x for x in range(1, 101)])
Ys = np.array(Xs*3)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(1, activation="linear", input_dim=1))
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=0.0001), loss="mean_squared_error", metrics=['accuracy'])
```

```python
history = model.fit(Xs, Ys, batch_size=10, epochs=300, validation_split=.2)
```

```python
model.predict([200.2223])
```

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
